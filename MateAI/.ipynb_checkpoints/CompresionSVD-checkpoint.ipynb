#### Cargamos los módulos necesarios

Como módulos necesarios simplemente __numpy__ para cálculos numéricos, __matplotlib__ para graficar, cargar y salvar imágenes a disco y __os__ para poder ver el tamaño que ocupan las imágenes (original y comprimida)

```python
from matplotlib.image import imread
from matplotlib.image import imsave
import matplotlib.pyplot as plt
import numpy as np
import os
```

#### Ahora cargamos la imagen y la mostramos

A partir de la imagen cargado original (color) tomamos el promedio de los tres canales para trabajar con esa nueva imagen en escala de grises.

```python
file ='mib-alien.jpg'
img_orig = imread(file)
A = np.mean(img_orig,-1)
img = plt.imshow(A)
img.set_cmap('gray')
plt.show()
```

#### Descomposición SVD

Realizamos la descomposición en valores singulares de la imagen. Luego eliminamos los datos correspondientes a valores singulares que no superen un porcentaje del máximo valor singular obtenido para la imagen original. 

```python
U,S,VT = np.linalg.svd(A)

# Obtengo la posición del valor singular que sea al menos el 10% del máximo
#--------------------------------------------------------------------------
S_max = np.amax(S)
umbral = 1/100            # 1%
S_thresh = umbral*S_max   # Elimino los datos relacionados con valores sigulares del 1% del máximo
S_cut = np.where(S < S_thresh)[0][0]

S=np.diag(S)

k = S_cut
k=200
A_new = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
img = plt.imshow(A_new)
img.set_cmap('gray')
plt.show()
imsave(file[:-4]+'_new.jpg', A_new, cmap='gray')
```

#### Porcentaje de compresión de la imagen
Evaluamos el tamaño de la imagen comprimida respecto de la imagen original

```python
size_orig = os.stat(file).st_size
size_new = os.stat(file[:-4]+'_new.jpg').st_size
reduccion = 100*(size_orig - size_new)/size_orig
print('El porcentaje de reducción es: {:.2f} %'.format(reduccion))
```

#### Distancia entre imágenes
Como medida de distancia entre imágenes analizamos el valor singular mayor de la diferencia entre las dos imagenes (original y comprimida)

```python
# Idea de la distancia entre imágenes
#------------------------------------
U,S,VT = np.linalg.svd(A-A_new)
print(np.amax(S))
```

#### Plus 1: Análisis de distancia
Analicemos qué pasa con la distancia entre imágenes (valor singular mayor de la diferencia) a medida que vamos eliminando valores singulares

```python
U_orig,S_orig,VT_orig = np.linalg.svd(A)
S_orig=np.diag(S_orig)
# Armamos un vector para ir recorriendo
#--------------------------------------
svds = np.arange(2, len(S))
dist = []
for k in svds:
    U = U_orig.copy()
    S = S_orig.copy()
    VT = VT_orig.copy()
    A_new = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
    U,S,VT = np.linalg.svd(A-A_new)
    dist = np.append(dist,np.amax(S))
    
```

```python
# Graficamos para visualizar mejor
#---------------------------------
fig = plt.figure()
plt.plot(svds, dist, marker='o', color='r')
plt.title('Distancia imagen original Vs. imagen comprimida')
plt.ylabel('Distancia')
plt.xlabel('# de valores singulares preservados en la compresión')
plt.show()
```

#### Conclusiones:

- Se puede ver que a medida que vamos preservando más (recortando menos) información relativa los valores singulares de menor valor tenemos una distancia entre imágenes menor.

- Este tipo de gráficas nos permitiría por ejemplo hasta que punto es aceptable recortar (o comprimir) la imagen para tener aún una distancia aceptable.

- A partir del gráfico anterior por ejemplo podríamos proponer un recorte a partir del valor singular 100. Una aproximación más conservadora podría ser a partir de el valor singular 200.

- Por supuesto, la definición de este corte se podrá basar en algún tipo de criterio y no a simple vista como propuesto en el párrafo anterior. Por ejemplo un promedio móvil sobre los datos de distancia evaluando la varianza en cada grupo. Se esperaría que al estabilizar la distancia al seguir dejando valores singulares la varianza sea mínima.

#### Plus 2: Análisis de compresión por canal
Lo que buscamos ver en este caso es si existe variación en cuanto al valor singular elegido (dado un determinado criterio) para realizar la compresión, para los distintos canales de color.

```python
# Me armo una función porque la voy a llamar algunas veces
#---------------------------------------------------------
def findSVcut(Img,umbral):
    U,S,VT = np.linalg.svd(Img)
    S_max = np.amax(S)
    S_thresh = umbral*S_max   
    SV_cut = np.where(S < S_thresh)[0][0]
    S=np.diag(S)
    return SV_cut, U, S, VT
```

```python
# Separamos los tres canales
Ablue = img_orig[:,:,0]
Agreen = img_orig[:,:,1]
Ared = img_orig[:,:,2]

umbral = 1/100   # 1%

# Calculo para el canal azul
#---------------------------
S_cut, U_b, S_b, VT_b = findSVcut(Ablue,umbral)
print('Valor singular de corte para el canal azul: {}'.format(S_cut))

# Calculo para el canal rojo
#---------------------------
S_cut, U_g, S_g, VT_g = findSVcut(Agreen,umbral)
print('Valor singular de corte para el canal verde: {}'.format(S_cut))

# Calculo para el canal azul
#---------------------------
S_cut, U_r, S_r, VT_r = findSVcut(Ared,umbral)
print('Valor singular de corte para el canal rojo: {}'.format(S_cut))
```

#### Conclusiones:
- Como era esperable los valores difieren (cada canal tiene información distinta, más o menos variable, más o menos compresible)
- Podría definirse como criterio elegir el índice de valor singular más alto entre los tres canales, y utilizar dicho índice para los otros dos. En nuestro caso esto sería para el canal rojo, índice 47.
- Claramente no podría comprimirse (recortarse) cada canal por separado ya que no podríamos unir los canales en una misma imagen color dado que nos quedarían matrices de distinto tamaño

#### Probamos comprimir la imagen en color según este criterio

```python
k = 47

# Canal azul: Reconstruyo, trunco y convierto entero de 8 bits (porque me quedan valores decimales)
#--------------------------------------------------------------------------------------------------
Ablue_new = U_b[:,:k] @ S_b[0:k,:k] @ VT_b[:k,:]
np.clip(Ablue_new, 0, 255, out=Ablue_new)
Ablue_new = Ablue_new.astype('uint8')

# Canal verde: Reconstruyo, trunco y convierto entero de 8 bits (porque me quedan valores decimales)
#---------------------------------------------------------------------------------------------------
Agreen_new = U_g[:,:k] @ S_g[0:k,:k] @ VT_g[:k,:]
np.clip(Agreen_new, 0, 255, out=Agreen_new)
Agreen_new = Agreen_new.astype('uint8')

# Canal rojo: Reconstruyo, trunco y convierto entero de 8 bits (porque me quedan valores decimales)
#--------------------------------------------------------------------------------------------------
Ared_new = U_r[:,:k] @ S_r[0:k,:k] @ VT_r[:k,:]
np.clip(Ared_new, 0, 255, out=Ared_new) 
Ared_new = Ared_new.astype('uint8')

# Uno los tres canales
#---------------------
A_bgr = np.dstack((Ablue_new,Agreen_new,Ared_new))

# NOTA: Es necesario truncar para evitar la aparición de artefactos en la imagen
```

```python
plt.imshow(A_bgr)
```

```python
# Comparamos los tamaños
#-----------------------
imsave(file[:-4]+'_color_new.jpg', A_bgr)
size_orig = os.stat(file).st_size
size_new = os.stat(file[:-4]+'_color_new.jpg').st_size
reduccion = 100*(size_orig - size_new)/size_orig
print('El porcentaje de reducción es: {:.2f} %'.format(reduccion))
```
